```python
#reading the dataset
import pandas as pd
df=pd.read_csv('AmazonSalesData.csv')
print(df)
```

```python
#shape of the dataset
df_shape=df.shape
print("shape of DataFrame:",df_shape)
```

```python
#analysing number of rows and column
df_rows,df_columns=df.shape
print("number of rows:",df_rows)
print("number of columns:",df_columns)
```

```python
#analysing names of columns
column_names=df.columns.tolist()
print("column_names:\n",column_names)
```

```python
#Dataset summary
summary_chart=df.describe()
print(summary_chart)
```

```python
#reading first and last few rows
first_few_rows=df.head()
last_few_rows=df.tail()
print("first few rows:\n",first_few_rows)
print("last few rows:\n",last_few_rows)
```

```python
#analysing a specific column
specific_column=df['Region']
print("Specific column:\n",specific_column)
```

```python
#Analysing specific subset of data from dataset
subset_data=df.loc[40:60,['Order ID','Total Revenue']]
print("subset of 40 to 60 data:\n",subset_data)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('AmazonSalesData.csv')
df.hist(figsize=(10, 10), bins=20)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
print(df['Unit Cost'].value_counts())
sns.countplot(x='Unit Cost', data=df)
plt.xlabel("X-values")
plt.ylabel("Y-values")           
plt.show()
```
